```python
#Click Through Rate

import numpy as np
import pandas as pd
from datetime import datetime
import random

date_range = pd.date_range(start = "01-01-2023", end = "31-12-2024")

MEDICAL_Adverts = [
    "Anticoagulation Care Device Advert",
    "Human Immunodeficiency Detection Machine Advert",
    "HIV Medicine Advert",
    "Topical Corticosteroids Cream Advert",
    "Asthma Inhaler Advert",
    "Meningitis B Vaccine Advert",
    "Seasonal Influenza Vaccine Advert",
    "Hepatitis Vaccine Advert"
]

New_Date = []

Medical_Ads_Daily = []

Impressions = []

Clicks = []

for d in range(0,len(date_range)):

    for e in range(0,len(MEDICAL_Adverts)):
        pick_ad = MEDICAL_Adverts[e]
        Medical_Ads_Daily.append(pick_ad)
        pick_date = date_range[d]
        New_Date.append(pick_date)
        Impression_Numb = random.randint(200000, 300000)
        Impressions.append(Impression_Numb)
        if Impression_Numb <= 200000:
            Clicks_Numb = random.randint(2000,12000)
            Clicks.append(Clicks_Numb)
        else:
            Clicks_Numb = random.randint(5000,20000)
            Clicks.append(Clicks_Numb)
            
Click_data = {
            "Dates": New_Date,
            "Medical_Advert": Medical_Ads_Daily,
            "Impressions": Impressions,
            "Clicks": Clicks
}

df_click_rate = pd.DataFrame(Click_data)

df_click_rate["Click-through Rate (%)"] = (df_click_rate["Clicks"]/df_click_rate["Impressions"])*100

df_click_rate.to_excel("Click-through_Rate.xlsx", index=False)

df_click_rate.describe()
```

```python
#ROI Marketing

import numpy as np
import pandas as pd
from datetime import datetime
import random

date_range = pd.date_range(start = "01-01-2022", end = "31-12-2024", freq = 'MS').strftime("%b %Y").tolist()

Marketing_Campaign = [
    "Anticoagulation Care Device Marketing Campaign",
    "Human Immunodeficiency Detection Marketing Campaign",
    "HIV Medicine Advert Marketing Campaign",
    "Topical Corticosteroids Cream Advert Marketing Campaign",
    "Asthma Inhaler Advert Marketing Campaign",
    "Meningitis B Vaccine Advert Marketing Campaign",
    "Seasonal Influenza Vaccine Advert Marketing Campaign",
    "Hepatitis Vaccine Advert Marketing Campaign"
]

New_date_C = []
Generate_sales_growth = []
Average_Organic_Sales_Growth = []
Marketing_Cost = []
Medical_Campaign_Monthly = []

for f in range(0,len(date_range)):

    for g in range(0,len(Marketing_Campaign)):
        pick_Campaign = Marketing_Campaign[g]
        Medical_Campaign_Monthly.append(pick_Campaign)
        pick_date = date_range[f]
        New_date_C.append(pick_date)
        Monthly_sales_growth = random.randint(40000, 68000)
        Generate_sales_growth.append(Monthly_sales_growth)
        Organic_Sales_Growth = random.randint(10000, 20000)
        Average_Organic_Sales_Growth.append(Organic_Sales_Growth)
        Cost = random.randint(5000, 7000)
        Marketing_Cost.append(Cost)
       
Campaign_data = {
            "Dates": New_date_C,
            "Medical_Campaign": Medical_Campaign_Monthly,
            "Monthly Sales Growth (£)": Generate_sales_growth,
            "Monthly Average Organic Sales Growth (£)": Average_Organic_Sales_Growth,
            "Marketing Cost (£)": Marketing_Cost 
}

df_ROI = pd.DataFrame(Campaign_data)

df_ROI["Market Campaign ROI (%)"] = ((df_ROI["Monthly Sales Growth (£)"]-df_ROI["Monthly Average Organic Sales Growth (£)"]-df_ROI["Marketing Cost (£)"])/df_ROI["Marketing Cost (£)"])*100

df_ROI.head(20)

df_ROI.describe()

#df_ROI.to_excel("Market_Campaign_ROI.xlsx", index=False)
```

```python
#Procurement Total Spend

import random
import pandas as pd
from datetime import datetime, timedelta
from faker import Faker

fake = Faker()

# --------------------------------------------------
# STATIC CHOCOLATE PRODUCT LIST (exactly 5 options)
# --------------------------------------------------
SUPPLIER = []

for i in range(10):
    Company_name = fake.company()
    SUPPLIER.append(Company_name)

# --------------------------------------------------
# RANDOM DATE GENERATOR WITHIN A YEAR
# --------------------------------------------------
def random_date():
    start_date = datetime(2023, 1, 1)
    random_days = random.randint(0, 730)
    d = start_date + timedelta(days=random_days)
    return d.strftime("%d-%b-%y")  # e.g., 07-Jul-22


# --------------------------------------------------
# DATA GENERATOR FUNCTION
# --------------------------------------------------
def generate_procurement_data(n_rows):
    data = {
        "Company Name": [],
        "Date": [],
        "Amount": []
    }

    for _ in range(n_rows):
        cn = random.choice(SUPPLIER)
        dt = random_date()

        # amount e.g., $3,430
        amount = f"£{random.randint(550, 4000):,}"

        data["Company Name"].append(cn)
        data["Date"].append(dt)
        data["Amount"].append(amount)
        
    return pd.DataFrame(data)

df_procurement = generate_procurement_data(2738)

#df_procurement.to_excel("Spend_Data.xlsx", index=False)
```

```python
#On Time Deliveries

date_range = pd.date_range(start = "01-01-2022", end = "31-12-2024")

Total_Deliveries_Daily = []
On_Time_Daily = []

for a in range(0,len(date_range)):
    Total_Deliveries = random.randint(70000,80000)
    On_Time = random.randint(65000,70000)

    if On_Time > Total_Deliveries:
        Total_Deliveries = On_Time
        On_Time = Total_Deliveries
        Total_Deliveries_Daily.append(Total_Deliveries)
        On_Time_Daily.append(On_Time)
    else:
        Total_Deliveries_Daily.append(Total_Deliveries)
        On_Time_Daily.append(On_Time)
    
Deliver_data = {                   
            "Date": date_range, 
            "Total Deliveries": Total_Deliveries_Daily,
            "On Time Deliveries": On_Time_Daily
}

df_delivery = pd.DataFrame(Deliver_data)

df_delivery["On Time Delivery Percentage (%)"] = (df_delivery["On Time Deliveries"]/df_delivery["Total Deliveries"])*100

#df_delivery.to_excel("Delivery_Data.xlsx", index=False)

df_delivery.describe()
```

```python
#Contract Leakage

import random
import pandas as pd
from datetime import datetime, timedelta

# --------------------------------------------------
# RANDOM DATE GENERATOR WITHIN A YEAR
# --------------------------------------------------
def random_date():
    start_date = datetime(2023, 1, 1)
    random_days = random.randint(0, 730)
    d = start_date + timedelta(days=random_days)
    return d.strftime("%d-%b-%y")  # e.g., 07-Jul-22

# --------------------------------------------------
# DATA GENERATOR FUNCTION
# --------------------------------------------------
def generate_leakage_data(n_rows):
    data = {
        "Date": [],
        "Leakage Amount (£)": [],
        "Non-Leakage Amount (£)": [],
        "Sum of Purchase Orders (£)": [],
        "Percentage Leakage (%)": []
    }

    for _ in range(n_rows):
        dt = random_date()

        # amount e.g., $3,430
        leakage_amount = random.randint(200,400) 
        non_leakage_amount = random.randint(1700,4500)
        Sum_PO = leakage_amount + non_leakage_amount
        leakage_percent = (leakage_amount/Sum_PO)*100

        data["Date"].append(dt)
        data["Leakage Amount (£)"].append(leakage_amount)
        data["Non-Leakage Amount (£)"].append(non_leakage_amount)
        data["Sum of Purchase Orders (£)"].append(Sum_PO)
        data["Percentage Leakage (%)"].append(leakage_percent)
        
        
    return pd.DataFrame(data)

df_leakage = generate_leakage_data(1800)

#df_leakage.to_excel("Value_Leakage.xlsx", index=False)

df_leakage.describe()
```

```python
#Saving vs Target Saving

import random
import pandas as pd
from datetime import datetime, timedelta

date_range = pd.date_range(start = "01-01-2022", end = "31-12-2024", freq = 'MS').strftime("%b %Y").tolist()

saving_target = [6500, 8000, 10500, 11000, 14000, 15750, 16800, 18450, 19000, 22000, 24000, 25500] 

saving_year2023 = []
saving_year2024 = []

for j in range(12):
    targets = saving_target[j] + 1000
    targets2 = saving_target[j] + 1500
    saving_year2023.append(targets)
    saving_year2024.append(targets2)

all_saving_target = saving_target + saving_year2023 + saving_year2024

saving_data = {
        "Date": date_range,
        "Monthly Baseline (£)": [],
        "Monthly Price (£)": [],
        "Monthly Saving (£)": [],
        "Saving Target (£)": all_saving_target
    }

for h in range(0,len(date_range)):
    
    Baseline = random.randint(200000, 300000)
    Price = random.randint(170000, 270000)

    if Baseline - Price < 4500:
        New_Baseline = random.randint(210000,220000)
        New_Price = random.randint(200000,210000)
        saving_data["Monthly Baseline (£)"].append(New_Baseline)
        saving_data["Monthly Price (£)"].append(New_Price)
        saving_data["Monthly Saving (£)"].append(New_Baseline - New_Price)
    elif Baseline - Price > 30000:
        Big_Baseline = random.randint(200000,210000)
        Small_Price = random.randint(185000,190000)
    
        saving_data["Monthly Baseline (£)"].append(Big_Baseline)
        saving_data["Monthly Price (£)"].append(Small_Price) 
        saving_data["Monthly Saving (£)"].append(Big_Baseline - Small_Price)
    else:
        saving_data["Monthly Baseline (£)"].append(Baseline)
        saving_data["Monthly Price (£)"].append(Price) 
        saving_data["Monthly Saving (£)"].append(Baseline - Price)

    
    #saving_data["Saving Target (£)"].append(saving_target)

    #Forcast_saving = random.randint(40000, 68000)
    #saving_data["Forcast Saving (£)"].append(Forcast_saving)

df_saving = pd.DataFrame(saving_data)

df_saving.to_excel("Saving_Data.xlsx", index=False)

df_saving.describe()
```
